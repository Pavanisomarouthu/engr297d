```python
#Random points as dataset
#each data is a tuple (x,y)
points = [(np.array([[4]]),np.array([[2]])),(np.array([[2]]),np.array([[4]]))]
d = 1
print(points)
```

```python
def F(w):
    value = 0
    for x,y in points:
        value += (x.dot(w) - y) ** 2
    
    
    return value/len(points)
```

```python
def dF(w):
    value = 0
    for x,y in points:
        value += 2*(x.dot(w)-y)*x
    return value/len(points)
```

```python
def gradientDescent(F,dF,d):
    w = np.zeros(d)
    l = 0.01 #step size
    for t in range(500):
        value = F(w)
        grad = dF(w)
        w = w - l*grad #Taking one step
        print('iteration {}, w = {}, F(q) = {}, F\'(w) = {}'.format(t,w,value,grad))
```

```python
gradientDescent(F,dF,d)
```
