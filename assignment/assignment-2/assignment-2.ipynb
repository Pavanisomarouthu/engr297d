```python
#Linear Regression
import numpy as np
import random
import math
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False 


true_w = np.array([3])
inp = []
output = []
d = len(true_w)
points = []

for i in range(10000):
    x = np.random.randn(d)
    y = true_w.dot(x)+np.random.randn()
    points.append((x,y))
    inp.append(x)
    output.append(y)

inp = np.asarray(inp)
plt.scatter(inp[:, 0], output, s=0.1, color='k', marker='o')
plt.plot([-3.3,3.3],[-10,10])
```
